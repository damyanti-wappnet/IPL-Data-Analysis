```python
index = df.loc['2020-01-01':'2020-01-31'].index

#separates the first month's date
month_1 = df.loc['2020-01-01':'2020-01-31']

# seperate Close price for first month
Close_1_month = df.loc['2020-01-01':'2020-01-31']['Close'] 

#Close data for first month
Close_1_month 
```

```python
# Linegraph showing Close price over days for first month

figure, axis = plt.subplots()
axis.set_xlabel('Date')
axis.set_ylabel('Closing Price')
axis.set_title('Closing Price Over Time in January-2020')
plt.tight_layout()
figure.autofmt_xdate()
axis.plot(index,Close_1_month)
```

```python
index = df.loc['2020-01-01':'2020-06-30'].index

 # separate dates for first six months
month_6 = df.loc['2020-01-01':'2020-06-30']

# Close price for first six months
Close_6_month = df.loc['2020-01-01':'2020-06-30']['Close']

#Close data for first six months
Close_6_month
```

```python
# Linegraph showing Close price over days for first six months

figure, axis = plt.subplots()
axis.set_xlabel('Date')
axis.set_ylabel('Closing Price')
axis.set_title('Closing Price Over Time from January-2020 to June-2020')
plt.tight_layout()
figure.autofmt_xdate()
axis.plot(index,Close_6_month)
```

```python
index = df.loc['2020-01-01':'2020-12-31'].index

# separate dates for first year
year_1 = df.loc['2020-01-01':'2020-12-31'] 

#separate close price for first year
Close_year_1 = df.loc['2020-01-01':'2020-12-31']['Close'] 
Close_year_1
```

```python
# Linegraph showing Close price over days for first year

figure, axis = plt.subplots()
axis.set_xlabel('Date')
axis.set_ylabel('Closing Price')
axis.set_title('Closing Price Over Time in 2020')
plt.tight_layout()
figure.autofmt_xdate()
axis.plot(index,Close_year_1)
```

```python

# Candlestick chart for first month

from datetime import datetime
import plotly.graph_objects as go

fig = go.Figure(data=[go.Candlestick(x=month_1,open=df['Open'],high=df['High'],low=df['Low'],close=df['Close'])])
fig.update_layout(
    title='Stock Price For January-2020',
    yaxis_title='Stock Data',
    xaxis_title = 'Date'
)
fig.show()
```
