    <li>High: Highest price the stock reached during that day</li>
    <li>Low: Lowest price the stock is traded on that day</li>
    <li>Volume: The total amount of stocks traded on that day</li>

```python
# calculate Daily return of stock to know the gain or loss per day
# df['Daily Return'] = (df['Close']- df['Open'])/ df['Open']

# pct_change() is used to calculate percentage of changes in the daily return
df['Daily Return'] = df['Close'].pct_change()*100
df['Daily Return'].hist(bins=50)


# df['Daily Return'].head()
# df['Daily Return'].max()
df['Daily Return'].min()
```

```python
#  Line chart visualizing the trends in the close prices of Microsoft (with filters for time period of 1m, 6m, YTD, 1y, all)

import plotly.express as px

df = px.data.stocks()
df = yf.download('MSFT', start="2020-01-01", end="2022-11-30")

figure = px.line(df, x=df.index, y='Close',title='Close price Analysis with Time Period Selectors')

figure.update_xaxes(
    rangeselector=dict(
        buttons=list([
            dict(count=1, label="1m", step="month", stepmode="backward"),
            dict(count=3, label="3m", step="month", stepmode="backward"),
            dict(count=6, label="6m", step="month", stepmode="backward"),
            dict(count=1, label="YTD", step="year", stepmode="todate"),
            dict(count=1, label="1y", step="year", stepmode="backward"),
            dict(step="all")
        ])
    )
)
figure.show()
```

```python
# Bar chart for open, close, high, low trends in Microsoft prices(with filters for time period of 1m, 6m, YTD, 1y, all)

figure = px.bar(df, x=df.index, y='Close',title='Close price Analysis with Time Period Selectors')

figure.update_xaxes(
    rangeselector=dict(
        buttons=list([
            dict(count=1, label="1m", step="month", stepmode="backward"),
            dict(count=3, label="3m", step="month", stepmode="backward"),
            dict(count=6, label="6m", step="month", stepmode="backward"),
            dict(count=1, label="YTD", step="year", stepmode="todate"),
            dict(count=1, label="1y", step="year", stepmode="backward"),
            dict(step="all")
        ])
    )
)
figure.show()
```

```python
# candlestick chart for open, close, high, low trends in Microsoft prices(with filters for time period of 1m, 6m, YTD, 1y, all)


from datetime import datetime
import plotly.graph_objects as go
# figure = px.candlestick(df, x=df.index, y='Close',title='Close price Analysis with Time Period Selectors')
figure = go.Figure(data=[go.Candlestick(x=df.index,open=df['Open'],high=df['High'],low=df['Low'],close=df['Close'])])

figure.update_xaxes(
    rangeselector=dict(
        buttons=list([
            dict(count=1, label="1m", step="month", stepmode="backward"),
            dict(count=3, label="3m", step="month", stepmode="backward"),
            dict(count=6, label="6m", step="month", stepmode="backward"),
            dict(count=1, label="YTD", step="year", stepmode="todate"),
            dict(count=1, label="1y", step="year", stepmode="backward"),
            dict(step="all")
        ])
    )
)
figure.show()
```

```python
# Correlation between columns
# Positive correlation: both variables increase in relation to each other
# negative correlation: means that as one variable decrease, the other increases

corrmat = df.corr()
corrmat
```

```python

f, ax = plt.subplots(figsize =(9, 8))
sns.heatmap(corrmat, ax = ax, cmap ="YlGnBu", linewidths = 0.1)
```

```python
#show linear relation between variables

df = df[['Open','High','Low','Close','Adj Close']]

sns.pairplot(df, kind="scatter")
plt.show()
```

```python
# calculate the rolling average of the 'Open' column in the dataframe over a 50-day and 200 days period.
# A rising moving average indicates that the security is in an uptrend, 
# A declining moving average indicates that it is in a downtrend.

# df['Moving Average'] = df['Open'].rolling(30).mean()
df['MA50'] = df['Open'].rolling(50).mean()
df['MA200'] = df['Open'].rolling(200).mean()
df['Open'].plot(figsize = (15,7), legend = True)
df['MA50'].plot(legend = True)
df['MA200'].plot(legend = True)
```

```python
#Market Capitalisation
df['Marketcap'] = df['Open'] * df['Volume']
figure = px.line(df, x = df.index, y= df["Marketcap"])
figure.show()
# df['Marketcap'].plot(label = 'Microsoft', figsize = (12,6))
# plt.title('Market Cap')
# plt.legend()
# plt.show()
```
